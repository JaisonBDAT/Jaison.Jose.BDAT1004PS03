Step 4. Year 2061? Do we really have data from this year? Create a function to fix it 
and apply it.

```python
# Reference: https://stackoverflow.com/questions/37766353/pandas-to-datetime-parsing-wrong-year
import datetime
def dateUpdate(date):
#     inRng = (dat >= start_date) & (dat <= end_date)
#     dat2 = dat
    if date.year>2021:
        updated_year=date.year-100
    else:
        updated_year=date.year
    return datetime.date(updated_year,date.month,date.day)
```

```python
data['Date'] = data['Date'].apply(dateUpdate)
```

# Question 6

Create a line plot showing the number of marriages and divorces per capita in the 
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import csv
```

```python
marriage = pd.read_csv('us-marriages-divorces-1867-2014.csv')
columns = marriage.columns.drop(['Marriages','Divorces','Population','Year'])
```

```python
fig,ax= plt.subplots()
for column in columns:
    ax.plot(marriage['Year'],marriage[column])
ax.set_title('US Marriages & Divorces',fontsize=15)
ax.set_xlabel('Year',fontsize=12)
ax.set_ylabel('Count',fontsize=12)
ax.legend()
plt.show()
```

# Question 7


Create a vertical bar chart comparing the number of marriages and divorces per 
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset=pd.read_csv('us-marriages-divorces-1867-2014.csv')
sub_plot=dataset[(dataset.Year==1900) | (dataset.Year==1950) | (dataset.Year==2000)]
sub_plot.plot(x='Year',y=['Marriages','Divorces'], kind="bar",figsize=(10,10))
plt.title('Marriages and Divorces per capita',fontsize=15)
plt.xlabel('Years',fontsize=12)
plt.ylabel('Count per capita',fontsize=12)
plt.show()
```

# Question 8


Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort 
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv('actor_kill_counts.csv')
data=data.sort_values(['Count'])
data = data.reset_index(drop=True)
data.plot.barh('Actor','Count',figsize=(10,10))
plt.title('Deadliest actors in Hollywood',fontsize=15)
plt.xlabel('Kill Count',fontsize=15)
plt.ylabel('Actor Name',fontsize=15)
plt.yticks([])
for index, value in enumerate(data['Count']):
    plt.text(value, index, data.loc[index,'Actor'])
plt.show()
```
